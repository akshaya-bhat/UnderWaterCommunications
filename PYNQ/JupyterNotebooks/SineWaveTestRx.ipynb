```python
from pynq.overlays.base import BaseOverlay

base = BaseOverlay("base.bit")
```

```python
from pynq.lib.arduino import Arduino_IO
from pynq.lib.arduino import Arduino_Analog
from pynq.lib.arduino import ARDUINO_GROVE_A1
from pynq.lib import Pmod_ADC, Pmod_DAC

adc = Pmod_ADC(base.PMODA)
```

```python
analog1 = Arduino_Analog(base.ARDUINO, ARDUINO_GROVE_A1)
```

```python
analog1.read_raw()[0]
```

```python
import time

xAdcVoltages = []
pmodAdcVoltages = []
stopTime = 0
runTime = 0
startTime = time.time() 
samples = 0
maxSamples = 260000

while samples < maxSamples:
    #pmodAdcVoltages.append(adc.read())
    #pmodAdcAnalogValues.append(adc.read_raw())
    #xAdcAnalogValues.append(analog1.read_raw()[0])    
    xAdcVoltages.append(analog1.read()[0])    
    samples += 1
    
stopTime = time.time()
runTime = stopTime - startTime
print('run time {} seconds'.format(runTime))
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.plot(xAdcVoltages)
plt.xlabel('Samples')
plt.ylabel('voltage')
plt.title('Rx Transducer w/ PYNQ DAC sinewave output - farthest(voltage)')
plt.show()
```
