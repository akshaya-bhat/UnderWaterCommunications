# Receive Clicking

```python
from pynq.overlays.base import BaseOverlay
from pynq.lib import Pmod_ADC
import pynq.lib.rgbled as rgbled 
import matplotlib.pyplot as plt
import numpy as np
import time
```

```python
ol = BaseOverlay("base.bit")
```

```python
adc = Pmod_ADC(ol.PMODA)
adc.reset()
led4 = rgbled.RGBLED(4)

noColor = 0x0
blue = 0x1
green = 0x2
red = 0x4
```

```python
pmodAdcVoltages = []

seconds = 10
stopTime = 0
runTime = 0
startTime = time.time() 

voltageThreshold = 0.008

led4.write(red)

while runTime < seconds:
    voltage = adc.read()
    pmodAdcVoltages.append(voltage)
    
    if pmodAdcVoltages[-1][-1] >= voltageThreshold:
        led4.write(green)

    stopTime = time.time()
    runTime = stopTime - startTime
    
stopTime = time.time()
runTime = stopTime - startTime
print('run time {} seconds'.format(runTime))

led4.off()
```

# Rx Transducer w/ Tx Transducer in Water

```python
plt.plot(pmodAdcVoltages)
plt.xlabel('Samples')
plt.ylabel('voltage')
plt.title('Reading from PMOD ADC (voltage)')
plt.show()
```
